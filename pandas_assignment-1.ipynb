##### 10. Write a Pandas program to add summation to a row of the dataframe.

```python
import pandas as pd
irisdata3=pd.read_csv("C:\\Users\\Gogula Vinay\\Desktop\\Innomatics labs\\Data Science\\lms assignment\\9. Pandas\\Pandas_assignment\\Pandas_day_1\\Iris.csv",usecols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
irisdata3
```
